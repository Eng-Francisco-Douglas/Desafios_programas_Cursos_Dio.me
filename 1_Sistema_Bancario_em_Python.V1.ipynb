```python
nome = "João"  # Nome do titular da conta
saldo = 0  # Saldo inicial da conta
depositos = []  # Lista para armazenar os valores dos depósitos
saques = []  # Lista para armazenar os valores dos saques

# Loop principal
while True:
    print("\n======= MENU =======")
    print("1. Depósito")
    print("2. Saque")
    print("3. Extrato")
    print("0. Sair")

    opcao = int(input("Digite a opção desejada: "))

    if opcao == 1:
        valor_deposito = float(input("Digite o valor do depósito: "))
        if valor_deposito > 0:
            saldo += valor_deposito
            depositos.append(valor_deposito)
            print("Depósito de R${:.2f} realizado.".format(valor_deposito))
        else:
            print("Valor inválido para depósito.")

    elif opcao == 2:
        valor_saque = float(input("Digite o valor do saque: "))
        if valor_saque > 0:
            if valor_saque <= 500 and len(saques) < 3:
                if saldo >= valor_saque:
                    saldo -= valor_saque
                    saques.append(valor_saque)
                    print("Saque de R${:.2f} realizado.".format(valor_saque))
                else:
                    print("Saldo insuficiente para saque.")
            else:
                print("Limite diário de saque excedido.")
        else:
            print("Valor inválido para saque.")

    elif opcao == 3:
        print("Extrato de {}:".format(nome))
        print("Depósitos:")
        for deposito in depositos:
            print(" - R${:.2f}".format(deposito))
        print("Saques:")
        for saque in saques:
            print(" - R${:.2f}".format(saque))
        print("Saldo atual: R${:.2f}".format(saldo))

    elif opcao == 0:
        print("Encerrando o programa...")
        break

    else:
        print("Opção inválida. Digite um número válido.")

print("Programa encerrado.")
```
