```python
from time import sleep

class ContaBancaria:
    def __init__(self, nome, saldo_inicial=0):
        """
        Classe que representa uma conta bancaria.
        
        Args:
            Nome (str): O nome do titular da conta:
            Saldo_inicial (float, optional): Saldo inicial da conta. O padrão e 0
        """
        
        self.saldo = saldo_inicial
        
    def depositar(self, valor):
        """
        Realizar um deposito na conta bancaria.
        
        Args:
            Valor (float): Valor a ser depositado.
        """
        self.saldo += valor
        print(f"Deposito de {valor}R$ realizado com sucesso.")
        
    def sacar(self, valor):
        """
        Realiza um saque na conta bancaria, se houver saldo suficiente.
        
        Args:
            Valor (float): Valor a ser sacado
        """
        if self.saldo >= valor:
            self.saldo -= valor
            print(f"Saque de {valor}R$ realizado com sucesso.")
        else:
            print("Saldo insuficiente para realizar o saque.")
            
    def consulta_saldo(self):
        """
        Consulta o saldo atual da conta Bancaria
        """
        print(f"Saldo atual: {self.saldo}R$")

def main():
    """Função principal do programa."""
    print("===== Sistema Bancario ======")


    # Criar uma nova conta bancaria
    nome = str(input("Digite o seu nome: "))
    saldo_inicial = float(input("Digite o Saldo inicial: R$ "))
    conta = ContaBancaria(nome, saldo_inicial)


    # Menu de Opções
    while True:
        print('')
        print('='*30)
        print("""Escolha uma opção:
        [1] Depositar
        [2] Sacar:
        [3] Consultar Saldo:
        [0] Sair""")


        opcao = input("Digite sua opção: ")

        if opcao == "1":
            valor = float(input("Digite o valor a depositar: "))
            conta.depositar(valor)
        elif opcao == "2":
            valor = float(input("Digite o valor a sacar: "))
            conta.sacar(valor)
        elif opcao == "3":
            conta.consultar_saldo()
        elif opcao == "0":
            print("Saindo do programa....")
            sleep(3)
            print("Programa Encerrado com Sucesso.")
            break
        else:
            print("Opção inválida. Tente novamente.")


if __name__ == "__main__":
    main()
```
